```python
#Importing the necessary packages
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
games = pd.read_csv("C:\\Users\\n\\Downloads\\board_games (1).csv")
games.head
```

```python
#Question 1:
from sklearn.tree import DecisionTreeRegressor

# Create a subset of data with only playtime and rating columns
subset = games[['playtime', 'rating']].copy()

# Convert playtime from minutes to hours
subset['playtime'] = subset['playtime'] / 60

# Split the data into features (X) and target (y)
X = subset[['playtime']]
y = subset['rating']

# Fit a decision tree regression model
reg_tree = DecisionTreeRegressor(random_state=0)
reg_tree.fit(X, y)

# Predict the rating for a game with a playtime of 2 hours (120 min)
playtime = 2  # in hours
predicted_rating = reg_tree.predict([[playtime]])

print("Predicted rating for a game with a playtime of 2 hours:", round(predicted_rating[0], 1))
```

```python
#Question 2
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Create a subset of data with only Party.Game and rating columns
subset = games[['Party.Game', 'rating']].copy()

# Convert Party.Game column to binary numeric values (1 for "Yes", 0 for "No")
subset['Party.Game'] = subset['Party.Game'].map({'Yes': 1, 'No': 0})

# Split the data into features (X) and target (y)
X = subset[['Party.Game']]
y = subset['rating']

# Fit a decision tree regression model
reg_tree = DecisionTreeRegressor(random_state=0)
reg_tree.fit(X, y)

# Plot the regression tree
plt.figure(figsize=(8, 8))
plot_tree(reg_tree, feature_names=['Party.Game'], filled=True)
plt.show()

# Predict the rating for a game categorized as a Party.Game
party_game = 1  # 1 for "Yes"
predicted_rating = reg_tree.predict([[party_game]])

print("Predicted rating for a game categorized as a Party.Game:", round(predicted_rating[0], 1))
```

```python
#Question 3
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Create a subset of data with only the selected variables and rating column
subset = games[['max_players', 'playtime', 'min_age', 'age_of_game', 'rating']].copy()

# Split the data into features (X) and target (y)
X = subset[['max_players', 'playtime', 'min_age', 'age_of_game']]
y = subset['rating']

# Fit a decision tree regression model
tree3 = DecisionTreeRegressor(random_state=0)
tree3.fit(X, y)

# Plot the regression tree
plt.figure(figsize=(10, 8))
plot_tree(tree3, feature_names=['max_players', 'playtime', 'min_age', 'age_of_game'], filled=True)
plt.show()
```
